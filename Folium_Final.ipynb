```python
import numpy as np
import pandas as pd
import folium
from folium.features import CustomIcon
```

```python
jeju_df = pd.read_csv('jeju_orem.csv',encoding='cp949')
```

```python
jeju_map = folium.Map(
    location =[33.362418504726705, 126.52938114577637],
    zoom_start = 10,
)

tooltip='Click'


for i in range(jeju_df.shape[0]):
    
    icon_image = 'mountain.png'
    icon = CustomIcon(
    icon_image,
    icon_size = (20, 20),
    popup_anchor = (30,-30))
    
    folium.Marker([jeju_df.iloc[i]['위도'],jeju_df.iloc[i]['경도']],
    popup = f'<div style="width:100px"><strong>{jeju_df.iloc[i]["오름명"]}<strong><br>\
    주차장 : {jeju_df.iloc[i]["주차장"]}<br>\
    화장실 : {jeju_df.iloc[i]["화장실"]}<br><div>',
    tooltip = tooltip,
    icon = icon
    ).add_to(jeju_map)
jeju_map
```

```python
jeju_map = folium.Map(
    location =[33.362418504726705, 126.52938114577637],
    zoom_start = 10,
)

tooltip='Click'

car_is = folium.FeatureGroup(name='주차장유').add_to(jeju_map)
car_is_none = folium.FeatureGroup(name='주차장무').add_to(jeju_map)


for i in range(jeju_df.shape[0]):
    if jeju_df.iloc[i]["주차장"] == "Y":
        folium.Marker([jeju_df.iloc[i]['위도'],jeju_df.iloc[i]['경도']],
        popup = f'<div style="width:100px"><strong>{jeju_df.iloc[i]["오름명"]}<strong><br>\
        주차장 : {jeju_df.iloc[i]["주차장"]}<br>\
        화장실 : {jeju_df.iloc[i]["화장실"]}<br><div>',
        tooltip = tooltip,
        ).add_to(car_is)
    else:   
        folium.Marker([jeju_df.iloc[i]['위도'],jeju_df.iloc[i]['경도']],
        popup = f'<div style="width:100px"><strong>{jeju_df.iloc[i]["오름명"]}<strong><br>\
        주차장 : {jeju_df.iloc[i]["주차장"]}<br>\
        화장실 : {jeju_df.iloc[i]["화장실"]}<br><div>',
        tooltip = tooltip,
        ).add_to(car_is_none)
folium.LayerControl(collapsed=False).add_to(jeju_map)
jeju_map
```
